# Декомпозиция темпов роста цен на недвижимость на эффекты спроса и предложения

## Краткое описание 

В данной работе мы решили проверить какие факторы - спроса или предложения оказывали основное влияние на движение цен на рынке недвижимости в тот или иной период. Динамика вкладов рассматриваемых факторов дает лучшее понимание текущей конъюнктуры рынка и фазы делового цикла, а также формирует определенное представление о трендах, наблюдающихся на всем рынке, так и на рынках отедльных субъектов с точки зрения текущего и ретроспективного анализа.

**Целью** исследования является декомпозиция изменения цены на недвижимость на факторы спроса и предложения

**Идея исследования:**

Теоретическая основа работы была взята из статьи Shapiro 2022. Элементы кода были заимствованны из телеграмм канала "Холодный расчет".

В исследовании предполагается, что для каждого рыночного сектора существует кривая предложения (положительный наклон) и кривая спроса (отрицательный наклон).
Кривая предложения описывается уравнением:
$$q_i = \sigma_ip_i + \alpha_i $$
Кривая спроса описывается уравнением:
$$p_i = -\delta_iq_i + \beta_i $$
Шоки предложения и спроса определяются как изменения в константах этих кривых ($\Delta\alpha$,$\Delta\beta$). Их можно представить вот в таком виде:
$$Supply~~shock:~~~~~~ \Delta\alpha_i = \varepsilon_i^s = (q_{i,t} - \sigma_ip_{i,t}) - (q_{i,t-1} - \sigma_ip_{i,t-1})$$

$$Demand~~shock:~~~~~~ \Delta\beta_i = \varepsilon_i^d = (\delta_iq_{i,t} + p_{i,t}) - (\delta_iq_{i,t-1} + p_{i,t-1})$$

Все эти уравнения в свою очередь могут быть представлены в виде структурной векторной авторегрессии:
$$A^iz_{i,t} = \sum_{j=1}^{N}A_j^iz_{i,t-j} + \varepsilon_{i,t}$$
где 
$z_i = \begin{bmatrix}
    q_i \\
    p_i
\end{bmatrix}$ , $A^i = \begin{bmatrix}
    1 & -\sigma_i \\
    \delta_i & 1 
\end{bmatrix}$ , $\varepsilon_i = \begin{bmatrix}
    \varepsilon_i^s \\
    \varepsilon_i^d
\end{bmatrix}$

Матрица $A$ выглядит таким образом, так как мы накладываем ограничения на знак (наклоны кривых спроса и предложения) для идентификации структурной векторной авторегрессии. Эпсилон - интересующие нас шоки спроса и предложения. Однако мы не можем пронаблюдать их непосредственно, так как в преобразованной форме модель выглядит следующим образом:

$$z_{i,t} = [A^i]^{-1}\sum_{j=1}^{N}A_j^iz_{i,t-j} + v_{i,t},$$
где мы можем непосредственно пронаблюдать только $v_{i,t}$ (вектор шоков цены и объемов), $v_i = \begin{bmatrix}
    v_i^q \\
    v_i^p
\end{bmatrix}$, который связан с вектором шоков и предложения следующим образом:
$$\varepsilon_{i,t} = A^iv_{i,t}$$

То есть мы получаем некоторое уравнение, где шоки спроса и предложения зависят от шоков объема и цены. Легко убедиться, что ограничения на знак углового коэффициента кривых спроса и предложения (матрица $A$) накладывают ограничения и на знак шоков спроса и предложения. Так, если шоки цен и объемов имеют одинаковые знаки, это указывает на шок спроса. То есть положительные (или отрицательные) шоки одновременно для цены и количества означают положительный (или отрицательный) шок спроса в данное время, при этом знак шока предложения остаётся неизвестным. Шок цен и объемов разных знаков указывают на шок предложения, то есть положительные (или отрицательные) шоки для цены в сочетании с отрицательными (или положительными) шоками для количества свидетельствуют о наличии отрицательного (или положительного) шока предложения, при этом знак шока спроса остаётся неопределённым. В целом данные выводы соотносятся с ожидаемой теорией. 

Однако встает вопрос: как оценить шоки цены и объема.
Обычно нас не интересуют сами значения шоков, а интересует оценка матрицы $Ai$, получив которую мы строим функции импульсных откликов. Здесь же обратная ситуация. Мы понимаем как устроена $Ai$ и нас в целом не интересует ее оценка а интересуют знаки шоков. Можно в целом сказать что все рассуждения выше приведены для того, чтобы показать, что, зная знаки шоков цен и объемов, можно сказать к чему они в итоге привели - шоку спроса или шоку предложения.

Оказывается, шоки цен и объемов можно получить из простых векторных авторегрессий предполагая, что цены зависят от своих лагов, а также от лагов объемов и наоборот.

После расчета остатков этих регрессий мы можем сказать к какому типу шока они однозначно привели - спроса или предложения. Если знаки одинаковые - то это шок спроса, если разный - предложения. При этом ранее было сказано, что если мы можем однозначно идентифицировать направление шока спроса  (предложения) в периоде t, то мы ничего не можем сказать о шоке предложения (спроса) в этот период (хотя мы не можем исключить его существование), однако причину изменения интересующей нас переменной мы будем связывать только с одним фактором (это явное допущение).

Таким образом, мы связываем изменение цены на недвижимость в том или ином регионе в каждом периоде либо с шоком спроса, либо с шоком предложения. Затем, на основе этих декомпозиций, мы можем построить агрегированный индекс для всей страны, либо же для федеральных округов с учетом веса данного региона в общем объеме продаж недвижимости. Так, если в регионе номер 1 в периоде t шок цен на недвижимость определялся фактором спроса, а в регионе номер 2 в периоде t шок цен на недвижимость определялся фактором предложения, а доля региона 1 в общем объеме стройки составляет 20% а региона 2 80%, то если эти два региона образуют целую страну то в периоде t изменение цен на недвижимость в этой стране в периоде t на 80% определялось факторами спроса и на 20% факторами предложения.

**Описание данных:**

Мы будем производить оценку для цен недвижимости на первичном и на вторичном рынке. Для этого нам понадобится 4 датасета:
- **ddu_price_sa** - цены на недвижимость на первичном рынке по регонам РФ
- **ddu_volume_sa** - объемы выдачи ипотеки на первичном рынке по регонам РФ
- **dkp_price_sa** - цены на недвижимость на вторичном рынке по регонам РФ
- **dkp_volume_sa** - объемы выдачи ипотеки на вторичном рынке по регонам РФ

*P.S. все ряды были заранее сезонно скорректированны по методике X13 ARIMA SEATS*

```python
# импорт нужных библиотек 
import pandas as pd
import os
import statsmodels.api as sm
import numpy as np
import statsmodels
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
import matplotlib.pyplot as plt

from datetime import date
from statsmodels.tsa.api import VAR

import warnings
warnings.filterwarnings("ignore")
```

```python
# красивое отображение нижней оси
def beautiful_xticks(ax, data):
    ticklabels = ['']*len(data.index)
    ticklabels[::4] = [item.strftime('%b %d') for item in data.index[::4]]
    ticklabels[::12] = [item.strftime('%b %d\n%Y') for item in data.index[::12]]
    ax.xaxis.set_major_formatter(ticker.FixedFormatter(ticklabels))
    #plt.gcf().autofmt_xdate()
```

## Расчеты

```python
# число лагов (вводим с самого начала)
lag = 3
```

### DDU

```python
# загрузка данных 
price_ddu = pd.read_excel('ddu_price_sa.xlsx')
price_ddu = price_ddu.set_index('date')
volume_ddu = pd.read_excel('ddu_volume_sa.xlsx')
volume_ddu = volume_ddu.set_index('date')
```

```python
# доля городов в первичке 
share_ddu = volume_ddu.iloc[1:, 1:] \
         .div(volume_ddu.iloc[1:, 1:] \
         .sum(axis=1), axis=0)
```

```python
# логарифмируем данные и берем первую разность
price_ddu = price_ddu.applymap(lambda x: np.log(x)).diff(periods=1).dropna()
volume_ddu = volume_ddu.applymap(lambda x: np.log(x)).diff(periods=1).dropna()
```

```python
dt_start = date(2016,2,29)
dt_end = date(2023,10,31)

df_res_q_ddu = pd.DataFrame(
    index = pd.date_range(start=dt_start, end=dt_end, freq='M')
    , columns=share_ddu.columns.to_list())

df_res_p_ddu = pd.DataFrame(
    index = pd.date_range(start=dt_start, end=dt_end, freq='M')
    , columns=share_ddu.columns.to_list())

df_decomp_ddu = pd.DataFrame(
    index = pd.date_range(start=dt_start, end=dt_end, freq='M')
    , columns=['Supply','Demand'])
```

```python
# считаем остатки VAR для каждого региона
for prod in share_ddu.columns:
    df_pq = pd.concat([
    price_ddu.loc[dt_start:dt_end, [prod]].rename(columns={prod:'P'})
    , volume_ddu.loc[dt_start:dt_end, [prod]].rename(columns={prod:'Q'})]
    , axis=1)
    model = VAR(df_pq)
    lag = lag
    results = model.fit(lag)
    df_res_q_ddu[prod] = results.resid.Q
    df_res_p_ddu[prod] = results.resid.P
```

```python
# взвешиваем влияние факторов спроса и предложения в изменение цены по объему
df_mask_ddu = df_res_q_ddu * df_res_p_ddu
df_dem_ddu = price_ddu.mask(df_mask_ddu < 0, 0) 
df_sup_ddu = price_ddu.mask(df_mask_ddu > 0, 0)
df_decomp_ddu['Demand'] = share_ddu.mul(df_dem_ddu).sum(axis=1)
df_decomp_ddu['Supply'] = share_ddu.mul(df_sup_ddu).sum(axis=1)
```
